luz = 100.80

agua = 80.35

```python
salario_cientista_de_dados = 10000.00
parcela_bmw = 2500.45
parcela_apartamento = 3500.50
veterinario_romeuzinho = 500.00
luz = 100.80
agua = 80.35
condominio = 800.44
gas = 150.65
total_gastos = sum([parcela_bmw,parcela_apartamento,veterinario_romeuzinho,luz,agua,condominio,gas])
valor_liquido = salario_cientista_de_dados - total_gastos

print("*"*95)
print("\nSeu salário de cientista de dados jr é de R$ {:.2f} e seus gastos estão dispostos a seguir:".format(salario_cientista_de_dados))
print('Parcela da Bmw: {}\n> Parcela do apartamento: {}\n> Veterinario do Romeuzinho: {}\n> Luz: {} \n> Água: {} \n> Condomínio: {} \n> Gas: {}\n'.format(parcela_bmw,parcela_apartamento,veterinario_romeuzinho,luz,agua,condominio,gas))

print("Total Restante: R${:.2f}\n".format(valor_liquido))
print("*"*95)
```
